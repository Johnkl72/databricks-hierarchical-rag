```python
from databricks.vector_search.client import VectorSearchClient
import time
```

```python
dbutils.widgets.text("catalog", "workspace")
dbutils.widgets.text("schema", "default")
dbutils.widgets.text("endpoint_name", "vector-search-demo")

# 2. Leer los valores
CATALOG = dbutils.widgets.get("catalog")
SCHEMA = dbutils.widgets.get("schema")
ENDPOINT_NAME = dbutils.widgets.get("endpoint_name")

# 3. Construir el nombre del √≠ndice autom√°ticamente
# IMPORTANTE: Debe coincidir con la l√≥gica del Notebook 1
INDEX_NAME = f"{CATALOG}.{SCHEMA}.pdf_hierarchical_index"

print(f"üîÑ Preparando sincronizaci√≥n para:")
print(f"   üìç Endpoint: {ENDPOINT_NAME}")
print(f"   üîé √çndice:   {INDEX_NAME}")
```

```python
print(f"üîÑ Iniciando sincronizaci√≥n de: {INDEX_NAME}")

vsc = VectorSearchClient()

try:
    # 1. Obtener el objeto √≠ndice
    index = vsc.get_index(ENDPOINT_NAME, INDEX_NAME)
    
    # 2. Enviar la orden de sincronizaci√≥n
    index.sync()
    
    print("üöÄ ¬°Se√±al de sincronizaci√≥n enviada exitosamente!")
    print("‚ÑπÔ∏è El √≠ndice procesar√° los cambios en segundo plano (As√≠ncrono).")
    print("‚úÖ Tarea finalizada.")

except Exception as e:
    print(f"‚ùå Error al contactar el √≠ndice: {e}")
    raise e # Esto har√° que el Job se ponga rojo si falla la conexi√≥n
```
